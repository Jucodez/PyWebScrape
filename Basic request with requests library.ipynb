First we have to get the data from the internet.<br>
This can be achieved with the request library using the get request
Data to be extracted is in json format.

```python
# Importing the request library
import requests
# Importing the pandas library for data analysis
import pandas as pd
```

```python
# assign the url of the data source to a variable
url= 'http://py4e-data.dr-chuck.net/comments_1815381.json'
```

```python
# use get request to extract data from url
r=requests.get(url)
```

```python
#check request status
# a status code of 200 means successful
r.status_code
```

```python
# get info about requested data and check for content -type
r.headers
```

```python
#since data in json format use .json() method to extract data
r.json()
```

```python
# store comments data in variable and transform to dataframe
data=r.json()['comments']
df=pd.DataFrame(data)
```

```python
#view dataframe
df
```

```python
# find sum of comments
df['count'].sum()
```
