```python
# Import the pandas library
import pandas as pd 
```

```python
# Specify the URL of the Wikipedia page containing the list of colleges and universities in Indiana.
url='https://en.wikipedia.org/wiki/List_of_colleges_and_universities_in_Indiana'


# Use the pandas function read_html to scrape tables from the given URL.
# The match parameter is used to specify a string that should be present in the table's title or attributes.
# This helps to filter out specific tables from the HTML page.
df=pd.read_html(url, match='List of active institutions')
```

```python
# The read_html function returns a list of DataFrames parsed from the HTML tables on the webpage.
# Since we used match='List of active institutions', the first DataFrame in the list is the one we are interested in.
# Extract the first DataFrame from the list.
df[0]
```
