```python
# Import required libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Set the URL to scrape data from
url = 'https://www.scrapethissite.com/pages/forms/'
```

```python
# Send a GET request to the URL
r = requests.get(url)
```

```python
# Check the status code of the response
print(r.status_code)
```

```python
# Get the HTML content from the response
html_text = r.text
```

```python
html_text
```

```python
# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(html_text, 'html.parser')
```

```python
# Find the HTML table element
table_html = soup.find('table')
```

```python
# Find all the table rows (tr elements)
tr_list = table_html.find_all('tr')
```

```python
tr_list
```

```python
# Print the text content of the first table header (th) cell
print(tr_list[0].find_all('th')[0].text.strip())
```

```python
# Create an empty list to store the extracted data
data = []

# Loop through the table rows and extract data
for i, tr in enumerate(tr_list):
    if i > 0:  # Skip the header row (i=0)
        datapoint = {}
        # Loop through the table data (td) cells in the current row
        for k, td in enumerate(tr.find_all('td')):
            # Get the corresponding header from the header row and use it as the key
            header = tr_list[0].find_all('th')[k].text.strip()
            datapoint[header] = td.text.strip()
        data.append(datapoint)

# Print the extracted data
print(data)
```

```python
# Create a pandas DataFrame from the extracted data
df = pd.DataFrame(data)
```
